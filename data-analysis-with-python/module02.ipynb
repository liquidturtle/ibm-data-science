```python
import pandas as pd
import numpy as np
df = pd.read_csv('automobile.csv')
df = df.replace("?", np.nan)
df.info()
```

```python
# add +1 on every value of symboling column
df['symboling'] = df['symboling']+1
df.head(5)
```

```python
# change column type to float and replace nan values with column mean
df['normalized-losses'] = df['normalized-losses'].astype(float)
df['normalized-losses'] = df['normalized-losses'].replace(np.nan, df['normalized-losses'].mean())
df.head()
```

```python
# find rows with missing price values
missing_price = df[df['price'].isna()]
missing_price.head(10)
```

```python
# drop rows with missing price values
df2 = df.dropna(subset=['price'], axis=0, inplace=False)
df2.info()
```

```python
#df2['price'] = df2['price'].astype(int)
df2['price'] = pd.to_numeric(df2['price'], downcast='integer')
df2.info()
```

```python
# convert mpg to lp100km
df2['city-mpg'] = (235/df2['city-mpg']).round(decimals=1)
df2.rename(columns={'city-mpg':'city-lp100km'}, inplace=True)
df2.head()
```

```python
# data normalization
df['length-simple-feat-scaling'] = df['length'] / df['length'].max()
df['length-min-max-scaling'] = (df['length'] - df['length'].min()) / (df['length'].max() - df['length'].min())
df['length-zscore'] = (df['length'] - df['length'].mean()) / df['length'].std()
df.sample(5)
```

```python
# data binning
df.dropna(subset=['price'], inplace=True)
df['price'] = pd.to_numeric(df['price'])
bins = np.linspace(min(df['price']), max(df['price']), 4)
print(bins)
group_names = ['Low', 'Medium', 'High']
df['price-binned'] = pd.cut(df['price'], bins, labels=group_names, include_lowest=True)
df.sample(20)
df["price-binned"].hist()
df.plot.scatter(x='length-zscore', y='price')
```

```python
# turn categorical into quantitative data with one hot encoding
fuel_dummies = pd.get_dummies(df['fuel-type'], prefix='fuel')
df = pd.concat([df, fuel_dummies], axis=1)

# one-liner
df = pd.concat([df, pd.get_dummies(df['make'])], axis=1)

df.head()
```
